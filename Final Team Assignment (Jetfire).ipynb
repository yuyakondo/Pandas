# Strategy for missing values and Outliers

To Begin, we realized that 374 data points were missing. To create accurate data, we came out with three different approaches to fill in missing values: by using secondary resources, using other countries' data, and applying mean and median from historical data.

First, we decided to research secondary resources to fill 2011 data results for each country variable. An example of this is Malaria cases in the European countries where according to the WHO report, we placed 0 because there were no malaria casualties in 2015 (WHO Europe, 2016). This can be attributed to the geographical attributes of the disease.

Even though we used multiple secondary resources, we noticed some small European countries are not independent. Our team assumed that these countries might already have their data in the central country's data to where they belongs and we extrapolated the population value for their central governments to create a ratio proportion to fill in missing values.

The second method was to use similar country situation and population size. For example, data was filled data if the countries have an identical political and population situation. 

Finally, missing values were imputed by mean and median application to the other regions data. 

# Finding Correlation

After figuring out the missing values, we found the correlation between every variable shown below to make it visualize the result of the correlation, we decided to use heatmaps. The following code is showing the correlation result: 

```python
# Showing the correlation of entire variables 

jetfire_df.corr(method = 'pearson').round(decimals = 2)
```

To make it visualize the result of the correlation, we decided to use the heatmap. The following code is showing the correlation result with heatmap. 

```python
# Heatmap for the entire columns of Western European countries

# converting correlation matrix into a DataFrame
jetfire_df_corr = jetfire_df.corr(method = 'pearson').round(decimals = 2)

# specifying plot size (making it bigger)
fig, ax = plt.subplots(figsize=(60,60))

sns.heatmap(data = jetfire_df_corr)

# developing a spicy heatmap
sns.heatmap(data       = jetfire_df_corr, # the correlation matrix
            cmap       = 'coolwarm',     # changing to COOLWARM colors
            square     = True,          # tightening the layout
            annot      = True,          # should there be numbers in the heatmap
            linecolor  = 'black',       # lines between boxes
            linewidths = 1)          # how thick should the lines be?


# title and displaying the plot
plt.title("""
Linear Correlation Heatmap for Western European Countries
""")

# Displaying the heatmap
plt.show()
```

Definition of five variables from World Bank: 

Maternal Mortality Ratio: - The maternal mortality ratio is the ratio of the number of maternal deaths during a given time period 100,000 live births during the same time-period. A maternal death refers to a female death from any cause related to or aggravated by pregnancy or its management (excluding accidental or incidental causes) during pregnancy and childbirth or within 42 days of termination of pregnancy, irrespective of the duration and site of the pregnancy.  

Antiretroviral therapy coverage: - standard antiretroviral therapy (ART) consists of the combination of antiretroviral (ARV) drugs to maximally suppress the HIV virus and stop the progression of HIV disease. ART also prevents onward transmission of HIV. Huge reductions have been seen in rates of death and infections when use is made of a potent ARV regimen, particularly in early stages of the disease.  

GDP per person employed is gross domestic product (GDP) divided by total employment in the economy. Purchasing power parity (PPP) GDP is GDP converted to 2011 constant international dollars using PPP rates. An international dollar has the same purchasing power over GDP that a U.S. dollar has in the United States.

Trade is the sum of exports and imports of goods and services measured as a share of gross domestic product.

Adjusted net enrollment is the number of pupils of the school-age group for primary education, enrolled either in primary or secondary education, expressed as a percentage of the total population in that age group.
